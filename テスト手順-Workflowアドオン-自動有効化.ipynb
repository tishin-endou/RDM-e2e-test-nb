## 自動有効化の結果を検証する

expect_auto_activate=Trueの場合: テンプレートが「有効化されたワークフロー」パネルに自動的に表示されること
expect_auto_activate=Falseの場合: テンプレートが「有効化されたワークフロー」パネルに表示されないこと

```python
async def _step(page):
    # Wait for panel to be visible
    activations_panel = page.locator('#activationsPanel')
    await expect(activations_panel).to_be_visible(timeout=transition_timeout)
    await activations_panel.scroll_into_view_if_needed()
    
    # Check if template is in the activated workflows list
    activated_template = activations_panel.locator(f'//td//strong[text()="{workflow_name}"]')
    
    if expect_auto_activate:
        # Template should be automatically activated
        await expect(activated_template).to_be_visible(timeout=transition_timeout)
        await activated_template.scroll_into_view_if_needed()
    else:
        # Template should NOT be automatically activated
        # First ensure the panel is loaded
        await expect(activations_panel).to_be_visible(timeout=transition_timeout)
        await expect(activated_template).to_have_count(0, timeout=transition_timeout)

await run_pw(_step)
```

## 「アドオン」をクリックし、Workflowアドオンのテンプレート一覧で対象行の無効化ボタンをクリックし、ダイアログで確認ボタンをクリックする
ステータスが「無効」に変わること

```python
async def _step(page):
    addons_link = page.locator('//a[text() = "アドオン"]')
    await addons_link.scroll_into_view_if_needed()
    await addons_link.click()
    header = page.locator('//h3[text() = "アドオンを選択"]')
    await expect(header).to_be_visible(timeout=transition_timeout)
    row = page.locator('#localWorkflowsPanel').locator(f'//tr[.//strong[text()="{workflow_name}"]]')
    await row.locator('//button[*[text() = "無効にする"]]').click()
    await expect(page.locator('//h4[text() = "ワークフローテンプレートの無効化"]')).to_be_visible(timeout=transition_timeout)
    await page.locator('//button[contains(@data-bind, "confirmDisableTemplate")]').click()
    await expect(row.locator('//*[contains(@class, "label") and text() = "無効"]')).to_be_visible(timeout=transition_timeout)
    await asyncio.sleep(3)

await run_pw(_step)
```
